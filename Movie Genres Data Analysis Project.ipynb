```python
movies_genre.head()
```

```python
movies_genre['genres'].str.split('|')
```

```python
from pandas import Series, DataFrame 
```

```python
split = movies_genre['genres'].str.split('|').apply(Series,1).stack()
split.index = split.index.droplevel(-1)
split.name = 'genres_split'
del movies_genre['genres']
movies_genre = movies_genre.join(split)
```

```python
movies_genre
```

```python
### Research Questions
#1. Which genres are the most common (number of movies made)?
genres_count = pd.DataFrame(movies_genre.groupby('genres_split').original_title.nunique()).sort_values('original_title', ascending = True)
```

```python
genres_count
```

```python
import matplotlib.pyplot as plt

# Horizontal bar chart with custom color
genres_count['original_title'].plot.barh(title='Movies Per Genre', color='#0097b2', figsize = (10,9))

plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.show()
```

```python
#2. Which genres have high avg. budget and revenue?
genres_avg = movies_genre.groupby('genres_split').mean(numeric_only=True)
pd.options.display.float_format = '{:2f}'.format

genres_avg
```

```python
movies_counted = (movies_genre['vote_count'] >= 50).astype(int)
movies_counted.corr(movies_genre['vote_average'], method='spearman')
```

```python
import matplotlib.pyplot as plt

# Filter movies with at least 50 votes
movies_counted = movies_genre[movies_genre['vote_count'] >= 50]

# Scatter plot of vote_average vs profit
movies_counted.plot.scatter(x='vote_average', y='profit', color='#0097b2', figsize=(8,6), title='Vote Average vs Profit')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter movies with at least 50 votes
movies_counted = movies_genre[movies_genre['vote_count'] >= 50]

# Create figure with size
plt.figure(figsize=(8,6))

# Regression plot
sns.regplot(x='vote_average', y='profit', data=movies_counted, color='#0097b2', line_kws = {"color": '#0d253f'})

# Add title
plt.title('Vote Average vs Profit')

# Show plot
plt.show()
```

```python
#5. Take a look at profit per genre per year
movies_genre.head()
```

```python
import pandas as pd

# Group by release year and genre, then calculate average profit
avg_profit = pd.DataFrame(
    movies_genre.groupby(['release_year', 'genres_split'])['profit'].mean()
).reset_index()

avg_profit
```

```python
final_genre = pd.pivot_table(avg_profit, values = 'profit', index = ['genres_split'], columns = ['release_year'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pivot table (you already have this)
final_genre = pd.pivot_table(avg_profit, values='profit', index='genres_split', columns='release_year')

# Create heatmap
plt.figure(figsize=(12,8))
sns.heatmap(final_genre, cmap='YlGnBu', linewidths=0.5)

plt.title('Genres by Profit per Year')
plt.xlabel('Release Year')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```
